<h3> Splitting </h3>

```python
X = df.iloc[:,[3,4]]
print(f"X Shape {X.shape}")
X.head()
```

<h2> Clustering using K-Means </h2>

<h3> Using a for loop we will build and train a K-means model where n_clusters ranging from 2 to 12. Each time the model is trained, we will calculate the inertia and add it to the list of inertia_errors, lastly we will calculate the silhouette score and add it to the list of silhouette_scores.</h3>

```python
#defining the clusters
n_clusters = range (2,13)
inertia_errors = [] #list
silhouette_scores = [] #list
#Adding a for loop to train the model and calculate the inertia, silhouette scores.
for k in n_clusters:
    model=KMeans(n_clusters = k, random_state=42, n_init =10)
    model.fit(X)
    inertia_errors.append(model.inertia_) #calculating inertia
    silhouette_scores.append(silhouette_score(X,model.labels_)) #adding the scores
print("Intertia: ", inertia_errors[:3])
print()
print("Silhouette SCores: ", silhouette_scores[:3])
```

<h2> Elbow Plot </h2>

```python
#line plot of inertia_erros vs n_clusters
x_values = list(range(2,13))

plt.figure(figsize=(8,6))
sns.set(style="whitegrid")

plt.plot(x_values, inertia_errors, marker='o', linestyle='-',color='b')

plt.title('K-Means Model: Inertia vs Number of Clusters')
plt.xlabel('No. of clusters')
plt.ylabel('Intertia')
plt.grid(True)
plt.show()
```

```python
#line plot of silhouette_scores vs n_clusters
x_values = list(range(2,13))

plt.figure(figsize=(8,6))
sns.set(style="whitegrid")

plt.plot(x_values, silhouette_scores, marker='o', linestyle='-',color='b')

plt.title('K-Means Model: Inertia vs Number of Clusters')
plt.xlabel('No. of clusters')
plt.ylabel('Silhouette Scores')
plt.grid(True)
plt.show()
```

<h3> The best number of clusters is 5 </h3>

```python
final_model = KMeans(n_clusters=5,random_state=42,n_init=10)
final_model.fit(X)
```

```python
labels = final_model.labels_
centroids = final_model.cluster_centers_
print(labels[:5])
print(centroids[:5])
```

```python
#plotting annual income vs spending scores with final_model labels
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], hue=labels, palette='deep')
sns.scatterplot(
    x = centroids[:,0],
    y = centroids[:,1],
    color = 'black',
    marker = '+',
    s =500
)
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Annual Income vs Spending Score")
```

<h3> Grouping </h3>

```python
xgb = X.groupby(final_model.labels_).mean()
```

```python
xgb
```

```python
#creating a side by side bar chart of the grouped data "xgb"
plt.figure(figsize = (8,6))

x = [0,1,2,3,4]
x_labels = labels
income_values = xgb['Annual Income (k$)']
spending_values = xgb['Spending Score (1-100)']

bar_width=0.35
index = range(len(x))

plt.bar(index, income_values, bar_width, label='Annual Income')
plt.bar([i + bar_width for i in index], spending_values, bar_width, label='Spending Score')

plt.xlabel("Clusters")
plt.ylabel("Value")
plt.title("Annual Income and Spending Score by Cluster")
plt.xticks([i + bar_width / 2 for i in index], x)
plt.legend()

plt.show()
```
